#Using GradientTape with constant tensor imputs

```python
input_const= tf.constant(3.)
with tf.GradientTape () as tape:
  tape.watch(input_const)
  result = tf.square(input_const)

gradient = tape.gradient(result, input_const)
print(gradient)
```

You squared 3 to get 9.

Then you asked: “If I nudge 3 slightly up or down, how fast does 9 change?”

Answer: 6 (because near 3, squaring changes pretty quickly).

-------
result = tf.square(input_const)
Computes:

result = 3² = 9

So inside the tape, it recorded the operation “square”.

gradient = tape.gradient(result, input_const)
Now you ask the tape:

“What’s the derivative of result with respect to input_const?”

Since result = x², the derivative is:

(x²)' = 2x

At x = 3, that’s: 2 * 3 = 6

✅ So gradient will be 6.0

## Difference between a tf.constant and a tf.Variable is:

tf.constant = a fixed number/tensor (TensorFlow treats it like “this won’t change”)

tf.Variable = a changeable number/tensor (made to be updated during training)

#Example: what is the acceleration?
Mesure the position of falling apple along a vertical axis over time.



```python
time = tf.Variable(0.)
with tf.GradientTape () as outer_tape:   # will compute "derivative of the derivative" (9.8t)' = 9.8
  with tf.GradientTape () as inner_tape: # will compute first derivative (4.9t^2)'=4.9*2*t
    position = 4.9 * time**2
  speed = inner_tape.gradient(position, time)
acceleration = outer_tape.gradient(speed, time)

print(speed)
print(acceleration)
print(time)
print(position)
```
